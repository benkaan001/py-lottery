
- There are some numbers that are drawn more frequently than others in Powerball. The five most frequently drawn numbers of all time are `32, 39, 22, 41, 36, and 16`, according to the Powerball statistics tracking website Powerball.net. <br>


- WHEN 1 TICKET PURCHASED PER DRAWING - 3 TICKETS TOTAL IN A WEEK.

```python
import random 
import json 

JACKPOT = 1_000_000_000
FIVE_WHITES_ONLY= 1_000_000
FOUR_WHITES_PB= 50_000
FOUR_WHITES_ONLY= 100
THREE_WHITES_PB = 100
THREE_WHITES_ONLY = 7
TWO_WHITES_PB = 7
ONE_WHITE_PB = 4
PB_ONLY = 4

LUCKY_WHITES = {32, 39, 22, 41, 36}
LUCKY_PB = 16

whiteball_possibilities = list(range(1, 70))
powerball_possibilities = list(range(1, 27))

tickets_purchased_per_drawing = 1
num_of_drawings_per_week = 3
num_of_drawings_per_year = 156

total_amount_spent = 0
total_amount_earned = 0 

winnings = {
    "JACKPOT": 0,
    "5": 0,
    "4+PB": 0,
    "4": 0,
    "3+PB": 0,
    "3": 0,
    "2+PB": 0,
    "1+PB": 0,
    "PB": 0,
    "NO WIN": 0
}

def calculate_total_winnings(ticket_numbers, winning_numbers):
    """compares the ticket_numbers played against winning_numbers, updates
    the winning dictionary, and returns the updated win amount
    """
    win_amount = 0

    # get the matching numbers between the winning_numbers set and ticket_numbers set
    # assingn the length of the set to save the number of times whiteballs matched
    white_matches = len(ticket_numbers["whites"].intersection(winning_numbers["whites"]))
    powerball_match = ticket_numbers['powerball'] == winning_numbers["powerball"]

    # check if jackpot
    if white_matches == 5:
        if powerball_match:
            win_amount = JACKPOT
            #increment the winnings dict
            winnings["JACKPOT"] += 1
        else:
            win_amount = FIVE_WHITES_ONLY
            # update/increment the winnings dict
            winnings["5"] += 1
    
    elif white_matches == 4:
        if powerball_match:
            win_amount = FOUR_WHITES_PB
            #increment the winnings dict
            winnings["4+PB"] += 1
        else:
            win_amount = FOUR_WHITES_ONLY
            # update/increment the winnings dict
            winnings["4"] += 1
    
    elif white_matches == 3:
        if powerball_match:
            win_amount = THREE_WHITES_PB
            #increment the winnings dict
            winnings["3+PB"] += 1
        else:
            win_amount = THREE_WHITES_ONLY
            # update/increment the winnings dict
            winnings["3"] += 1

    elif white_matches == 2 and powerball_match:
        win_amount = TWO_WHITES_PB
        #increment the winnings dict
        winnings["2+PB"] += 1
    
    elif white_matches == 1 and powerball_match:
        win_amount = ONE_WHITE_PB
        #increment the winnings dict
        winnings["1+PB"] += 1
    
    elif powerball_match:
        win_amount = PB_ONLY
        winnings["PB"] += 1

    else:
        winnings["NO WIN"] += 1


    return win_amount



for drawing in range(num_of_drawings_per_week):
    whiteball_drawing = set(random.sample(whiteball_possibilities, k=5))
    powerball_drawing = random.choice(powerball_possibilities)

    winning_numbers = {"whites": whiteball_drawing, "powerball": powerball_drawing}

    for ticket in range(tickets_purchased_per_drawing):
        total_amount_spent += 2
        # switch random ticket values to the LUCKY numbers
        # whiteball_values_chosen = set(random.sample(whiteball_possibilities, k=5))
        # powerball_value_chosen = random.choice(powerball_possibilities)
        ticket_numbers = {"whites": LUCKY_WHITES, "powerball": LUCKY_PB}
     
        # calculate the win amount per ticket
        win_amount = calculate_total_winnings(ticket_numbers, winning_numbers)
        # update the total earnings
        total_amount_earned += win_amount

        print(f"winning_numbers: {winning_numbers}")
        print(f"ticket_numbers: {ticket_numbers}")

print(f"Amount spent on tickets: ${total_amount_spent}")
print(f"Amount earned from winnings: ${total_amount_earned}")
# print the winnings dictionary 
print(json.dumps(winnings, indent=2))




```
